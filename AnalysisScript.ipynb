```python
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter
```

```python
fnames =['Farhan_192122_26_02_2022','Javed_190001_06_02_2022','Rameez_192122_21_02_2022','Sehrish_190001_04_02_2022','Zahid_192122_07_02_2022','owais_192122_26_01_2022','yasir_192122_29_01_2022','FullTranscription']
for y in range(0,len(fnames)):
  tfile=open(f"./{fnames[y]}.txt")
  t2array=tfile.readlines()
  sentence_length =[]
  name_list=[]
  audio_dur =[]
  tlines=[]
  for x in range(0,len(t2array)):
      line=t2array[x].strip(' ').split()
      tlines.append(line[2:])
      sentence_length.append(len(line[2:]))
      audio_dur.append(float(line[1])-float(line[0]))
      name_list.append(f"L{x+1}")
  plt.figure(dpi =300)
  plt.title(f"{fnames[y]}")
  plt.xlabel("Audio duration in seconds")
  plt.ylabel("Number of words")
  plt.scatter(audio_dur,sentence_length)
  plt.savefig(f"{fnames[y]}.png")
  plt.show()


```
